# インポート

```python
from sklearn.datasets import load_iris

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
iris = pd.read_csv("iris.csv")
iris.head(2)
```

# plt直下を直接使う（あんまり使わない）

## 折れ線グラフ

```python
plt.plot(iris.sep_len)
```

## 散布図

```python
plt.scatter(iris.sep_len, iris.sep_wid)
```

# 複数の絵を描く方法

```python
fig = plt.figure(figsize=(14, 7))
ax1 = fig.add_subplot(121)
ax1.scatter(iris.sep_len, iris.sep_wid)
ax2 = fig.add_subplot(122)
ax2.scatter(iris.pet_len, iris.pet_wid)
```

## 散布図（種類によって色を変える）

```python
for i in iris.target.unique():
    temp = iris.loc[iris.target==i,:]
    plt.scatter(temp.sep_len, temp.sep_wid)
```

# Seabornを使う

## 散布図

```python
sns.scatterplot(x="sep_len", y="sep_wid", hue="target", data=iris)
```

## 棒グラフ

```python
sns.barplot(x="target", y="sep_len", data=iris)
```

## 複数のグラフを描く

```python
fig = plt.figure(figsize=(14, 7))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
sns.scatterplot(x="sep_len", y="sep_wid", hue="target", data=iris, ax=ax1)
sns.barplot(x="target", y="sep_len", data=iris, ax=ax2)
```
